```python
import pandas as pd
import numpy as np
```

```python
# Membaca dataset
df = pd.read_csv('data_tugas 2.csv')

# Menampilkan 5 data pertama
df.head()
```

```python
# Mengecek jumlah data hilang
missing = df.isnull().sum()

# Menampilkan hasil
print(missing)
```

```python
kategori = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']
```

```python
df[kategori] = df[kategori].fillna('unknown')
```

```python
for col in kategori:
    print("\nKolom:", col)
    print(df[col].value_counts())
```

```python
numerik = ['length', 'width', 'height']
```

```python
rata_rata = df[numerik].mean()
print(rata_rata)
```

```python
df[numerik] = df[numerik].fillna(rata_rata)
```

```python
df['wheel-bases-normalized'] = (
    (df['wheel-bases'] - df['wheel-bases'].min()) /
    (df['wheel-bases'].max() - df['wheel-bases'].min())
)

# Menampilkan hasil
df[['wheel-bases', 'wheel-bases-normalized']].head()
```

```python
for col in numerik:
    df[col + '_zscore'] = (df[col] - df[col].mean()) / df[col].std()

# Menampilkan hasil
df[['length_zscore', 'width_zscore', 'height_zscore']].head()
```

```python
df.to_csv('hasil_data_preparation.csv', index=False)
```
